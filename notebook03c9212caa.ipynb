Challenge 1: Word Frequency Counter

**Hint:**
* Use split() to break the paragraph into words.
* Use a dictionary where the key is the word and the value is the frequency of that word.

```python
# Input: "hello world hello everyone"
# Output: {'hello': 2, 'world': 1, 'everyone': 1}
```

```python
txt = "hello world hello everyone"
```

```python
x = txt.split()
print (x)
```

```python
my_dict={}
```

```python
for item in x:
    if item in my_dict:
        my_dict[item] +=1
    else:
        my_dict[item] =1
```

```python
print(my_dict)
```
